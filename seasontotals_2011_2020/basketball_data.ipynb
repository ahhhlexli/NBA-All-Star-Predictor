```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('bball_stats.csv', index_col='name')
df.head()
```

```python
pd.set_option('display.max_columns', None)
df.describe()
```

```python
df.head()
```

```python
df.positions = df.positions.str.extract(r'.*\.(\w+):')
df.team = df.team.str.extract(r'.*\.(\w+)')
df.head()
```

```python
minbyage = df.groupby(df.age)['minutes_played'].mean().plot()
#df.plot(minbyage)
#df.plot.bar(x='age', y='minutes_played')
```

```python
df.head()
```

```python
df.groupby('positions')['points'].mean().plot(rot=45)
```

```python
df.sort_values(by='points', ascending=False)
points_df = df.sort_values(by='points', ascending=False)[:50]
#points_df
```

```python
points_df['points_off_twos'] = (points_df['made_field_goals'] - points_df['made_three_point_field_goals']) * 2
#points_df['points_off_twos']
```

```python
points_df.groupby('positions')['made_free_throws', 'points_off_threes', 'points_off_twos', 'points'].mean()
```

```python
points_df['points_off_threes'] = points_df['made_three_point_field_goals'].apply(lambda x: x*3)
points_df.groupby('positions')[ 'made_free_throws', 'points_off_threes', 'points_off_twos'].sum().plot.bar(stacked=True,rot=45)
```
