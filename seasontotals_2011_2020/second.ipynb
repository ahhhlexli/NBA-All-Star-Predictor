```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns; sns.set()
sns.set(rc={'figure.figsize':(8, 6)})
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('bball_stats.csv')
df['rebounds'] = df.offensive_rebounds + df.defensive_rebounds
pd.set_option('display.max_columns', None)
```

```python
two_point_total = df.points_off_twos.sum()
three_point_total = df.points_off_threes.sum()
ft_total = df.made_free_throws.sum()
```

```python
#line graph of year on the x and % on the y
season_df = df.groupby('season').sum().reset_index()
```

```python
season_df
```

```python
#season_df['twopoint_percentage'] = season_df.apply(two_percentage(season_df.points, season_df.points_off_twos), axis=1)
season_df['twopoint_percentage'] = round(season_df.points_off_twos / season_df.points, 3) *100
season_df['threepoint_percentage'] = round(season_df.points_off_threes / season_df.points, 3) *100
season_df['ft_percentage'] = round(season_df.made_free_throws / season_df.points, 3) *100
season_df
```

```python
p1 = sns.lineplot(data=season_df[['twopoint_percentage', 'threepoint_percentage', 'ft_percentage']])
#p1.set_xticklabels(labels=season_df.season)
p1.set_xticks(range(len(season_df.season)))
p1.set_xticklabels(list(range(2011, 2021)))
p1.set_title('Breakdown of Overall Points Distribution 2011-2020')
p1.set(xlabel='Season', ylabel='Percentage', xlim=(-0.5,9.5), ylim=(0,100))
p1.legend(labels=['Two Point %', 'Three Point %', 'Free Throw %'])
```

```python
fig, ax = plt.subplots(1,1)

sns.lineplot(data=season_df, ax=ax, x='season', y='twopoint_percentage', color='Slateblue', linewidth=3, marker='o')
sns.lineplot(data=season_df, ax=ax, x='season', y='threepoint_percentage', color='mediumseagreen', linewidth=3, marker='')
sns.lineplot(data=season_df, ax=ax, x='season', y='ft_percentage', color='Firebrick', linewidth=3, marker='o')

ax.set_title('Breakdown of Overall Points Distribution 2011-2020')
ax.set(xlabel='Season', ylabel='Percentage',  ylim=(0,100), xlim=(2011, 2020))
ax.legend(labels=['Two Point %', 'Three Point %', 'Free Throw %'])
```

```python
#find top point scorers overall in the 10 year period
top = df.groupby('name')['points'].sum()
plist = list(top.sort_values(ascending=False)[:10].index)
```

```python
pdf = df.groupby(['season', 'name']).sum().reset_index('season')
top20_df = pdf.loc[plist].reset_index()
#sns.lineplot(data=cropped_df[['twopoint_percentage', 'threepoint_percentage', 'ft_percentage']])
```

```python
top20_df['twopoint_percentage'] = round(top20_df.points_off_twos / top20_df.points, 3) *100
top20_df['threepoint_percentage'] = round(top20_df.points_off_threes / top20_df.points, 3) *100
top20_df['ft_percentage'] = round(top20_df.made_free_throws / top20_df.points, 3) *100
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, figsize=(20,10))

sns.lineplot(data=top20_df,ax=ax1, x='season', y='twopoint_percentage', hue='name', legend=False)
ax1.set_title('Percentage of Points Scored from Two-Point Range 2011-2020')
ax1.set(xlabel='Season', ylabel='Percentage',  ylim=(20,90), xlim=(2011, 2020))
#ax1.legend(loc=5,borderaxespad=-12)

sns.lineplot(data=top20_df,ax=ax2, x='season', y='threepoint_percentage', hue='name')
ax2.set_title('Percentage of Points Scored from Three-Point Range 2011-2020')
ax2.set(xlabel='Season', ylabel='Percentage',  ylim=(0,60), xlim=(2011, 2020))
ax2.legend(loc=5,borderaxespad=-13)

sns.lineplot(data=top20_df,ax=ax3, x='season', y='ft_percentage', hue='name', legend=False)
ax3.set_title('Percentage of Points Scored from Freethrows 2011-2020')
ax3.set(xlabel='Season', ylabel='Percentage',  ylim=(5,40), xlim=(2011, 2020))
#ax3.legend(loc=5,borderaxespad=-12)

sns.lineplot(data=season_df, ax=ax4, x='season', y='twopoint_percentage', color='Slateblue', linewidth=3, marker='o')
sns.lineplot(data=season_df, ax=ax4, x='season', y='threepoint_percentage', color='mediumseagreen', linewidth=3, marker='')
sns.lineplot(data=season_df, ax=ax4, x='season', y='ft_percentage', color='Firebrick', linewidth=3, marker='o')
ax4.set_title('Breakdown of Overall Points Distribution 2011-2020')
ax4.set(xlabel='Season', ylabel='Percentage',  ylim=(0,70), xlim=(2011, 2020))
ax4.legend(labels=['Two Point %', 'Three Point %', 'Free Throw %'], loc=5, borderaxespad=-10)
```

```python
team_df = df.groupby(['team', 'season']).sum().reset_index()

team_df['twopoint_percentage'] = round(team_df.points_off_twos / team_df.points, 3) *100
team_df['threepoint_percentage'] = round(team_df.points_off_threes / team_df.points, 3) *100
team_df['ft_percentage'] = round(team_df.made_free_throws / team_df.points, 3) *100

team_df['att_3point_percentage'] = round(team_df.attempted_three_point_field_goals / team_df.attempted_field_goals, 3) *100
team_df['att_2point_percentage'] = round((team_df.attempted_field_goals - team_df.attempted_three_point_field_goals) / team_df.attempted_field_goals, 3) *100
```

```python
sns.lineplot(data=team_df, x='season', y='att_3point_percentage')
sns.lineplot(data=team_df, x='season', y='att_2point_percentage')
```
