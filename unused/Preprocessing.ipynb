```
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```
df = pd.read_csv('/content/drive/MyDrive/basketball_reference_scrape/all_data.csv')
```

```
df.head()
```

```
df['position'] = df['position'].str.extract(r'(\w+),?')
```

```
df['position'].value_counts()
```

```
df.groupby('position')['all_star'].value_counts()
```

```
df.columns
```

To Drop: Player, Season, League

Categorical Types: Team, Position

Keep: EFG, 3pt%, 2pt%, FT%

Continuous: Everything else


```
cat_cols = ['team', 'position']

for column in cat_cols:
  temp = pd.get_dummies(df[column], drop_first=True, prefix=column)
  df = pd.concat([df, temp], axis=1)
  del df[column]
```

```
df = df.drop(columns=['player', 'season', 'league'], axis=1)
```
